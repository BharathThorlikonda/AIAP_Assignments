```python
class ShoppingCart:
    def __init__(self):
        self.items = {}

    def add_item(self, name, price):
        if not isinstance(name, str) or not name:
            return "Invalid input: Item name must be a non-empty string."
        if not isinstance(price, (int, float)) or price < 0:
            return "Invalid input: Item price must be a non-negative number."

        if name in self.items:
            self.items[name]['quantity'] += 1
        else:
            self.items[name] = {'price': price, 'quantity': 1}
        return f"Added {name} to the cart."

    def remove_item(self, name):
        if not isinstance(name, str) or not name:
            return "Invalid input: Item name must be a non-empty string."

        if name in self.items:
            if self.items[name]['quantity'] > 1:
                self.items[name]['quantity'] -= 1
            else:
                del self.items[name]
            return f"Removed one {name} from the cart."
        else:
            return f"{name} not found in the cart."

    def total_cost(self):
        total = 0
        for item in self.items.values():
            total += item['price'] * item['quantity']
        return round(total, 2)
```

Now, let's create some test cases for the `ShoppingCart` class.

```python
# Test cases for ShoppingCart

# Test Case 1: Add items and calculate total cost
cart1 = ShoppingCart()
print("Test Case 1: Add items and calculate total cost")
print(cart1.add_item("Apple", 0.5))
print(cart1.add_item("Banana", 0.3))
print(cart1.add_item("Apple", 0.5))
print(f"Total cost: {cart1.total_cost()}")
print("-" * 20)

# Test Case 2: Remove an item
cart2 = ShoppingCart()
cart2.add_item("Orange", 1.0)
cart2.add_item("Grape", 2.5)
cart2.add_item("Orange", 1.0)
print("Test Case 2: Remove an item")
print(cart2.remove_item("Orange"))
print(f"Total cost: {cart2.total_cost()}")
print("-" * 20)


# Test Case 3: Remove an item not in the cart
cart3 = ShoppingCart()
cart3.add_item("Pear", 1.2)
print("Test Case 3: Remove an item not in the cart")
print(cart3.remove_item("Mango"))
print(f"Total cost: {cart3.total_cost()}")
print("-" * 20)

# Test Case 4: Add item with invalid name
cart4 = ShoppingCart()
print("Test Case 4: Add item with invalid name")
print(cart4.add_item("", 1.0))
print(cart4.add_item(123, 1.0))
print("-" * 20)

# Test Case 5: Add item with invalid price
cart5 = ShoppingCart()
print("Test Case 5: Add item with invalid price")
print(cart5.add_item("Pineapple", -2.0))
print(cart5.add_item("Pineapple", "free"))
print("-" * 20)

# Test Case 6: Remove item with invalid name
cart6 = ShoppingCart()
cart6.add_item("Kiwi", 0.8)
print("Test Case 6: Remove item with invalid name")
print(cart6.remove_item(""))
print(cart6.remove_item(456))
print("-" * 20)

# Test Case 7: Total cost of an empty cart
cart7 = ShoppingCart()
print("Test Case 7: Total cost of an empty cart")
print(f"Total cost: {cart7.total_cost()}")
print("-" * 20)
```
