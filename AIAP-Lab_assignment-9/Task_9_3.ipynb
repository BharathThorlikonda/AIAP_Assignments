# **Manually written**

```python
"""
Simple Calculator Script
------------------------
This script defines basic arithmetic functions such as addition,
subtraction, multiplication, and division.

Each function contains a NumPy-style docstring that explains
parameters, returns, and examples.
"""

def add(a, b):
    """
    Add two numbers.

    Parameters
    ----------
    a : float or int
        First number.
    b : float or int
        Second number.

    Returns
    -------
    float
        The sum of a and b.

    Examples
    --------
    >>> add(10, 5)
    15
    """
    return a + b


def subtract(a, b):
    """
    Subtract one number from another.

    Parameters
    ----------
    a : float or int
        The number from which another number is subtracted.
    b : float or int
        The number to subtract.

    Returns
    -------
    float
        The result of a - b.

    Examples
    --------
    >>> subtract(10, 5)
    5
    """
    return a - b


def multiply(a, b):
    """
    Multiply two numbers.

    Parameters
    ----------
    a : float or int
        First number.
    b : float or int
        Second number.

    Returns
    -------
    float
        The product of a and b.

    Examples
    --------
    >>> multiply(4, 5)
    20
    """
    return a * b


def divide(a, b):
    """
    Divide one number by another.

    Parameters
    ----------
    a : float or int
        Numerator.
    b : float or int
        Denominator.

    Returns
    -------
    float
        The result of a / b.

    Raises
    ------
    ValueError
        If the denominator `b` is zero.

    Examples
    --------
    >>> divide(10, 2)
    5.0
    """
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b


# ---- Example usage ----
if __name__ == "__main__":
    # Demonstrate the calculator functions
    print("Addition:", add(10, 5))
    print("Subtraction:", subtract(10, 5))
    print("Multiplication:", multiply(10, 5))
    print("Division:", divide(10, 5))
```

# **AI Written**

```python
"""
Calculator Module

This module provides basic arithmetic operations: addition, subtraction,
multiplication, and division. Each function is designed for clarity and
includes examples for ease of use.
"""

def add(a, b):
    """
    Adds two numbers and returns the sum.

    Parameters
    ----------
    a : float or int
        The first number.
    b : float or int
        The second number.

    Returns
    -------
    float
        The sum of `a` and `b`.

    Examples
    --------
    >>> add(5, 3)
    8
    """
    return a + b


def subtract(a, b):
    """
    Subtracts the second number from the first.

    Parameters
    ----------
    a : float or int
        The number to subtract from.
    b : float or int
        The number to be subtracted.

    Returns
    -------
    float
        The result of the subtraction (`a` - `b`).

    Examples
    --------
    >>> subtract(10, 4)
    6
    """
    return a - b


def multiply(a, b):
    """
    Multiplies two numbers and returns the product.

    Parameters
    ----------
    a : float or int
        The first number.
    b : float or int
        The second number.

    Returns
    -------
    float
        The product of `a` and `b`.

    Examples
    --------
    >>> multiply(6, 7)
    42
    """
    return a * b


def divide(a, b):
    """
    Divides the first number by the second.

    Parameters
    ----------
    a : float or int
        The numerator.
    b : float or int
        The denominator.

    Returns
    -------
    float
        The result of the division (`a` / `b`).

    Raises
    ------
    ValueError
        If the denominator `b` is zero.

    Examples
    --------
    >>> divide(20, 5)
    4.0
    """
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b
```
