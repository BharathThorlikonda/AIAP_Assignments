```python
# Define a class to represent a student in SR University
class SRU_Student:
    # The constructor (__init__) is used to initialize the student's attributes
    def __init__(self, name, roll_no, hostel_status):
        # Store the student's name
        self.name = name

        # Store the student's roll number
        self.roll_no = roll_no

        # Store whether the student is a hosteller or day scholar
        self.hostel_status = hostel_status

        # Initialize the student's total fee as 0
        self.fee = 0

    # Method to update the student's fee
    def fee_update(self, amount):
        # Add the given amount to the student's total fee
        self.fee += amount
        # Print confirmation message
        print(f"Fee updated successfully! Current total fee: ₹{self.fee}")

    # Method to display all student details
    def display_details(self):
        # Print student's name
        print(f"Student Name: {self.name}")
        # Print student's roll number
        print(f"Roll Number: {self.roll_no}")
        # Print student's hostel status
        print(f"Hostel Status: {self.hostel_status}")
        # Print total fee paid so far
        print(f"Total Fee Paid: ₹{self.fee}")


# --- Main part of the program starts here ---

# Create an object (instance) of SRU_Student class
student1 = SRU_Student("Bharath Thorlikonda", "2505B03101", "Hosteller")

# Update the student's fee
student1.fee_update(50000)

# Display the student's complete details
student1.display_details()
```
