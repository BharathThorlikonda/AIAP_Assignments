# **Task Description#1 Basic Docstring Generation**

```python
def sum_even_odd(numbers):
    """
    Returns the sum of even and odd numbers in the given list.

    Parameters:
        numbers (list): A list of integers.

    Returns:
        tuple: A tuple (even_sum, odd_sum) where
               even_sum is the sum of all even numbers,
               and odd_sum is the sum of all odd numbers.
    """
    even_sum = 0
    odd_sum = 0

    for num in numbers:
        if num % 2 == 0:
            even_sum += num
        else:
            odd_sum += num

    return even_sum, odd_sum


# Example usage:
nums = [10, 15, 20, 25, 30]
even_sum, odd_sum = sum_even_odd(nums)
print("Sum of even numbers:", even_sum)
print("Sum of odd numbers:", odd_sum)
```
