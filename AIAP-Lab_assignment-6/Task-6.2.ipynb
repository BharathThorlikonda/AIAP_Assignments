# **Task Description#2 (loops)**


```python
def print_multiples(number):
  """Prints the first 10 multiples of a given number."""
  print(f"Multiples of {number}:")
  for i in range(1, 11):
    print(number * i)

# Example usage:
print_multiples(5)
```

```python
def print_multiples_while(number):
    i = 1
    while i <= 10:
        print(f"{number} x {i} = {number * i}")
        i += 1

print_multiples_while(7)
```
