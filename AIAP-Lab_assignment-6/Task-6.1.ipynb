# **Task Description#1 (Classes)**


```python
class Student:
    """Represents a student with a name, age, and major."""

    def __init__(self, name, age, major):
        """Initializes a new Student object."""
        self.name = name
        self.age = age
        self.major = major

    def get_details(self):
        """Returns a string with the student's details."""
        return f"Name: {self.name}, Age: {self.age}, Major: {self.major}"

# Create an instance of the Student class
student1 = Student("Bharath", 24, "AIAP")

# Test the get_details method
print(student1.get_details())
```
