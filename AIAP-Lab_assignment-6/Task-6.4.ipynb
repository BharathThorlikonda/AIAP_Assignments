# **Task Description#4 (For and While loops)**


```python
def sum_to_n(n):
  """Calculates the sum of the first n numbers using a for loop.

  Args:
    n: An integer representing the upper limit.

  Returns:
    The sum of the first n numbers.
  """
  total = 0
  for i in range(1, n + 1):
    total += i
  return total

# Example usage:
n = 10
result = sum_to_n(n)
print(f"The sum of the first {n} numbers is: {result}")
```

**Code Analysis:**

The `sum_to_n(n)` function calculates the sum of the first `n` positive integers.

1.  **Initialization:** It initializes a variable `total` to 0. This variable will store the cumulative sum.
2.  **Iteration:** It uses a `for` loop to iterate through a sequence of numbers from 1 to `n` (inclusive). The `range(1, n + 1)` function generates this sequence.
3.  **Accumulation:** In each iteration, the current number `i` is added to the `total`.
4.  **Return Value:** After the loop finishes, the function returns the final value of `total`, which is the sum of the first `n` numbers.

This approach is clear and efficient for calculating the sum using a loop.

**Alternative using a `while` loop:**

```python
def sum_to_n_while(n):
  """Calculates the sum of the first n numbers using a while loop.

  Args:
    n: An integer representing the upper limit.

  Returns:
    The sum of the first n numbers.
  """
  total = 0
  i = 1
  while i <= n:
    total += i
    i += 1
  return total

# Example usage:
n = 10
result_while = sum_to_n_while(n)
print(f"The sum of the first {n} numbers using a while loop is: {result_while}")
```

**Code Analysis (while loop):**

The `sum_to_n_while(n)` function achieves the same result as the `for` loop version but uses a `while` loop:

1.  **Initialization:** It initializes `total` to 0 and a counter variable `i` to 1.
2.  **Condition:** The `while i <= n:` condition checks if the counter `i` is less than or equal to `n`. The loop continues as long as this condition is true.
3.  **Accumulation and Increment:** Inside the loop, the current value of `i` is added to `total`, and then `i` is incremented by 1 (`i += 1`). This ensures that the loop eventually terminates when `i` becomes greater than `n`.
4.  **Return Value:** The function returns the final `total`.

Both the `for` loop and `while` loop examples demonstrate controlled looping to calculate the sum. The `for` loop is often preferred when the number of iterations is known beforehand, while the `while` loop is more suitable when the loop's continuation depends on a condition.
