# **Task Description#3 (Conditional Statements)**


```python
def classify_age_chained(age):
  """Classifies age into different groups using chained if-elif-else statements."""
  if age < 0:
    return "Invalid age"
  elif age < 13:
    return "Child"
  elif age < 20:
    return "Teenager"
  elif age < 60:
    return "Adult"
  else:
    return "Senior"

# Example usage:
print(classify_age_chained(8))
print(classify_age_chained(18))
print(classify_age_chained(45))
print(classify_age_chained(80))
print(classify_age_chained(-5))
```
