# **Task Description#2(loop)**

```python
# Faulty Code (Infinite Loop)
# i = 0
# while i < 5:
#     print(i)

# âœ… Corrected Code
i = 0
while i < 5:
    print(i)
    i += 1  # AI added increment step to prevent infinite loop
```

**Explanation of the corrected code for the Infinite Loop:**

The original faulty code was an infinite loop because the condition `i < 5` would always be true. The loop variable `i` started at 0 and the code inside the loop printed `i` but never changed the value of `i`. Without an increment or decrement step that would eventually make the condition `i < 5` false, the loop would continue to run indefinitely.

The corrected code fixes this by adding the line `i += 1` inside the `while` loop.

- `i += 1` is shorthand for `i = i + 1`.
- This line increments the value of `i` by 1 in each iteration of the loop.
- As a result, `i` will take on the values 0, 1, 2, 3, and 4.
- When `i` becomes 5, the condition `i < 5` is no longer true, and the loop terminates.

This ensures that the loop runs a finite number of times (in this case, 5 times) and prevents the program from getting stuck in an infinite loop.
