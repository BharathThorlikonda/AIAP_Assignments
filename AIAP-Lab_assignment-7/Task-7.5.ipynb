```python
# Faulty Code (Index Error)
# numbers = [1, 2, 3]
# print(numbers[5])

# âœ… Corrected Code
numbers = [1, 2, 3]

try:
    print(numbers[5])
except IndexError:
    print("Error: Index out of range! Showing safe access instead.")
    print("Last valid element:", numbers[-1])
```

**Explanation of the corrected code for Index Error:**

The original faulty code attempted to access an index (5) that was outside the valid range of the `numbers` list, which only has indices 0, 1, and 2. This would result in an `IndexError`.

The corrected code uses a `try...except` block to handle this potential error:

- The code inside the `try` block is executed first. This is where the potentially problematic operation, accessing `numbers[5]`, is attempted.
- If an `IndexError` occurs because the index is out of range, the code inside the `except IndexError:` block is executed.
- In the `except` block, a user-friendly error message "Error: Index out of range! Showing safe access instead." is printed.
- It then demonstrates "safe access" by printing the last valid element of the list using `numbers[-1]`. This provides a useful alternative or fallback when the requested index is invalid.

This approach prevents the program from crashing due to an `IndexError` and provides a way to gracefully handle such situations, informing the user about the error and potentially offering a safe alternative action.
