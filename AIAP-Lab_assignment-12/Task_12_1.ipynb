**Task: Write python code for linear_search () function to search a value in a list and extract its index.**

```python
def linear_search(data_list, search_value):
    """
    Performs a linear search to find the index of a value in a list.

    Args:
        data_list (list): The list to search through.
        search_value: The value to search for.

    Returns:
        int: The index of the search_value if found, otherwise -1.
    """
    for index, element in enumerate(data_list):
        if element == search_value:
            return index
    return -1

# Example usage:
my_list = [10, 4, 8, 1, 9, 23, 7]
value_to_find = 9

result_index = linear_search(my_list, value_to_find)

if result_index != -1:
    print(f"Value {value_to_find} found at index: {result_index}")
else:
    print(f"Value {value_to_find} not found in the list.")

value_to_find_not_present = 5
result_index_not_present = linear_search(my_list, value_to_find_not_present)

if result_index_not_present != -1:
    print(f"Value {value_to_find_not_present} found at index: {result_index_not_present}")
else:
    print(f"Value {value_to_find_not_present} not found in the list.")
```
