```python
import pandas as pd

def stats_from_csv(path, column):
  """
  Calculates the mean, min, and max of a numeric column in a CSV file.

  Args:
    path: The path to the CSV file.
    column: The name of the numeric column.

  Returns:
    A tuple containing the mean, min, and max of the column.
  """
  try:
    df = pd.read_csv(path)
    values = df[column].astype(float)
    mean_val = values.mean()
    min_val = values.min()
    max_val = values.max()
    return mean_val, min_val, max_val
  except FileNotFoundError:
    print(f"Error: File not found at {path}")
    return None, None, None
  except KeyError:
    print(f"Error: Column '{column}' not found in the CSV.")
    return None, None, None
  except ValueError:
    print(f"Error: Column '{column}' is not numeric.")
    return None, None, None

# Example usage with the uploaded file
file_path = '/content/sample_numbers.csv'  # Use the path where the file was saved
column_name = 'value'
mean_val, min_val, max_val = stats_from_csv(file_path, column_name)

if mean_val is not None:
  print(f"\nStatistics for column '{column_name}' from '{file_path}':")
  print("Mean:", mean_val)
  print("Min:", min_val)
  print("Max:", max_val)
```
