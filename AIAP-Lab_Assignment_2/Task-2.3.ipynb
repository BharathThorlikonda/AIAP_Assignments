```python
# Task 3 – Function to be explained by Gemini

import math

def area_of_shape(shape: str, **kwargs):
    """
    Returns the area of the requested shape.
    shape: one of 'circle', 'rectangle', 'triangle'
    For circle: radius
    For rectangle: width, height
    For triangle: base, height
    """
    shape = shape.lower()

    if shape == 'circle':
        r = kwargs.get('radius')
        if r is None:
            raise ValueError("Circle requires 'radius'")
        return math.pi * (r ** 2)

    elif shape == 'rectangle':
        w = kwargs.get('width')
        h = kwargs.get('height')
        if w is None or h is None:
            raise ValueError("Rectangle requires 'width' and 'height'")
        return w * h

    elif shape == 'triangle':
        b = kwargs.get('base')
        h = kwargs.get('height')
        if b is None or h is None:
            raise ValueError("Triangle requires 'base' and 'height'")
        return 0.5 * b * h

    else:
        raise ValueError(f"Unknown shape: {shape}")


# --- Test the function ---
print("Circle:", area_of_shape('circle', radius=5))
print("Rectangle:", area_of_shape('rectangle', width=4, height=6))
print("Triangle:", area_of_shape('triangle', base=8, height=5))
```
