# **Task 2 – Palindrome**

```python
# Task 2 – Palindrome Check (Gemini / Copilot style)

def is_palindrome(text: str) -> bool:
    """
    Check if 'text' is a palindrome, ignoring non-alphanumeric characters and case.
    """
    import re
    cleaned = re.sub(r'[^A-Za-z0-9]', '', text).lower()
    return cleaned == cleaned[::-1]

# --- Test the function ---
samples = [
    "A man, a plan, a canal: Panama",
    "Madam",
    "Hello",
    "Was it a car or a cat I saw?",
    "12321",
    "Python"
]

for s in samples:
    print(f"'{s}' → {is_palindrome(s)}")
```
